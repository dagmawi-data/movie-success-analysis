```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned/cleaned_movies.csv")
df.head()
```

```python
plt.scatter(df['IMDB_Rating'], df['Gross'])
plt.xlabel("IMDb Rating")
plt.ylabel("Gross Revenue")
plt.title("IMDb Rating vs Revenue")
plt.show()
# There is no strong direct relationship between IMDb rating and revenue.
# Some highly rated movies earn less, while some lower-rated movies earn very high revenue.
```

```python
top_directors = df.groupby('Director')['Gross'].mean().sort_values(ascending=False).head(10)

top_directors.plot(kind='bar')
plt.title("Top Directors by Average Revenue")
plt.ylabel("Average Gross")
plt.show()
# Certain directors consistently produce high-grossing movies,
# suggesting that director reputation may influence commercial success.
```

```python
plt.scatter(df['Runtime'], df['IMDB_Rating'])
plt.xlabel("Runtime (minutes)")
plt.ylabel("IMDb Rating")
plt.title("Runtime vs Rating")
plt.show()
# There is no clear relationship between runtime and IMDb rating.
# Both short and long movies can receive high ratings.
```

```python
plt.scatter(df['No_of_Votes'], df['IMDB_Rating'])
plt.xlabel("Number of Votes")
plt.ylabel("IMDb Rating")
plt.title("Votes vs Rating")
plt.show()
# Movies with more votes tend to have more stable ratings,
# but a high number of votes does not always mean a higher rating.
```

```python
yearly = df.groupby('Released_Year')['IMDB_Rating'].mean()

yearly.plot()
plt.title("Average IMDb Rating Over Years")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.show()
# Average IMDb ratings remain relatively stable over the years,
# with slight fluctuations but no strong upward or downward trend.
```
