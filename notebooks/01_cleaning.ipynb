```python
import pandas as pd

df = pd.read_csv(r"C:\Users\wwwww\Desktop\movie-success-analysis\data\raw\imdb_top_1000.csv")
df.head()
```

```python
# Check shape and info
print(df.shape)
print(df.info())

# Check missing values
print(df.isnull().sum())
```

```python

# 1. Convert Released_Year to numeric (integers) for proper analysis
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')

# 2. Split Genre into a list of genres for multi-label analysis
df['Genre'] = df['Genre'].str.split(', ')

# 3. Standardize Certificate: remove dashes, convert to uppercase
df['Certificate'] = df['Certificate'].str.replace('-', '', regex=False).str.upper()

# 4. Ensure Gross is numeric (remove $ or , if needed)
df['Gross'] = df['Gross'].replace('[\$,]', '', regex=True).astype(float)

# 5. Clean Director and Star names: strip extra spaces
for col in ['Director', 'Star1', 'Star2', 'Star3', 'Star4']:
    df[col] = df[col].str.strip()

# 6. Remove duplicate rows if any exist
df = df.drop_duplicates()

# 7. Final verification: check data types and first few rows
print(df.info())
print(df.head())
```

```python
#Released_Year: fill missing
df['Released_Year'] = df['Released_Year'].fillna(df['Released_Year'].median())
```

```python
#remove $ sign & convert to float
df['Gross'] = df['Gross'].replace('[\$,]', '', regex=True).astype(float)
```

```python
#Drop unusable columns
df = df.drop(columns=['Genre'])
```

```python
#Check for duplicates
df = df.drop_duplicates()
```
